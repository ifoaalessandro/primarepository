```python
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
df=pd.read_csv("https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/Startup.csv",sep=",")
```

```python
df
```

```python
X =df.drop(columns='Profit')
y =df['Profit']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.25, 
                                                    random_state = 667
                                                    )
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = model.predict(X_test)
```

```python
res_df = pd.DataFrame(data=list(zip(y_pred, y_test)),columns=['predicted', 'real'])
res_df
```

```python
res_df['error'] = res_df['real'] - res_df['predicted']
res_df
```

```python
res_df['error'].mean()
```

```python
length = y_pred.shape[0] #  
x = np.linspace(0,length,length)

plt.figure(figsize=(10,7))
plt.plot(x, y_test, label='real y')
plt.plot(x, y_pred, label="predicted y'")
plt.legend(loc=2);
```

```python
#misuro l'errore del mio modello
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error 
r2score= r2_score(y_test, y_pred)
mae=mean_absolute_error(y_test, y_pred)
mse=mean_squared_error(y_test, y_pred)
rmse=mean_squared_error(y_test,y_pred,squared=False)

print('R2_score:', r2score)
print('MAE:',mae)
print('MSE:',mse)
print('RMSE',rmse)
```
